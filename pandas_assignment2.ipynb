Group by and Aggregating

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Car/Flavors.csv")
```

```python
df
```

```python
group_by_frame=df.groupby('Base Flavor').mean(numeric_only=True)
group_by_frame
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
df.groupby(['Flavor Rating','Texture Rating']).agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
df.groupby('Base Flavor').describe()
```

###Merge,join and Concatenate


```python
import pandas as pd
```

```python
df1=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Car/LOTR.csv")
df1
```

```python
df2=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Car/LOTR 2.csv")
df2
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
### Inner jon
df1.merge(df2,how='inner',on=['FellowshipID','FirstName'])
```

```python
### Outer join
df1.merge(df2,how='outer',on=['FellowshipID','FirstName'])
```

```python
###Left join
df1.merge(df2,how='left',on=['FellowshipID','FirstName'])
```

```python
###Right join
df1.merge(df2,how='right',on=['FellowshipID','FirstName'])
```
